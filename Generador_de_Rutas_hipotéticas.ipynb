# Función para crear el lienzo en blanco

```python
#Importar librerías
import matplotlib.pyplot as plt
def diagramaenblanco(fig_width, fig_height,
                  bg_color="ivory", color="darkkhaki"):
    # Crear lienzo para la figura 16cmx12cm
    fig = plt.figure(figsize=(fig_width / 2.54, fig_height / 2.54))
    # Axes: posición izquierda 0, inferior 0, ancho 1, alto 1.(left, bottom, width, height)
    ax = fig.add_axes((0, 0, 1, 1))
    # Límites de la figura
    ax.set_xlim(0, fig_width)
    ax.set_ylim(0, fig_height)
    #Fondo
    ax.set_facecolor(bg_color)
    #Borrar ejes
    ax.tick_params(bottom=False, top=False,
                   left=False, right=False)
    ax.tick_params(labelbottom=False, labeltop=False,
                   labelleft=False, labelright=False)
    #Color de margen
    ax.spines["top"].set_color(color)
    ax.spines["bottom"].set_color(color)
    ax.spines["left"].set_color(color)
    ax.spines["right"].set_color(color)
    ax.spines["top"].set_linewidth(2)
    ax.spines["bottom"].set_linewidth(2)
    ax.spines["left"].set_linewidth(2)
    ax.spines["right"].set_linewidth(2)
    return fig, ax
```

# Caso 1: Sólido subenfriado a vapor sobrecalentado a diferente presión:
